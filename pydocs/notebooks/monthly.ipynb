## GRACE gravity field solutions

```python
# Imports
from gfeatpy.gravity import GravityField, EquivalentWaterHeight
from gfeatpy import plot as plt
```

```python
# Load static gravity field
static = GravityField("gravity/static/ITSG-Grace2018s.gfc", 90, 90)
```

### Non-resonant case

```python
# Load monthly static gravity field
monthly = GravityField("gravity/monthly/ITSG-Grace2018_n96_2007-03.gfc", 90, 90)
# Remove static field
monthly.coefficients = monthly.coefficients - static.coefficients
```

```python
# Plot EWH
plt.synthesis(monthly, 360, 180, EquivalentWaterHeight(0))
plt.show()
```

```python
# Plot EWH + Gaussian spatial smoothing filter r=750 km
plt.synthesis(monthly, 360, 180, EquivalentWaterHeight(750e3))
plt.show()
```
