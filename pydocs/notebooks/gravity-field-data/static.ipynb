## Static gravity field data

```python
# Imports
from gfeatpy.gravity import GravityField, GravityAnomaly, GeoidHeight
from gfeatpy import plotting
import matplotlib.pyplot as plt

data_root = "../../../data"
```

```python
# Load static gravity field
goco05c = GravityField(720).load(f"{data_root}/gravity/static/GOCO05c.gfc")
```

```python
# Plot gravity anomalies
contour, cbar = plotting.synthesis(goco05c, 1440, 720, GravityAnomaly(), z_max=200)
plt.show()
```

```python
# Plot geoid heights
plotting.synthesis(goco05c, 720, 360, GeoidHeight())
plt.show()
```
