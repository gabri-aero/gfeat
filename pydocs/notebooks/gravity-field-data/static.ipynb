## Static gravity field data

```python
# Imports
from gfeatpy.gravity import GravityField, GravityAnomaly, GeoidHeight
from gfeatpy import plot as plt

data_root = "../../../data"
```

```python
# Load static gravity field
goco05c = GravityField(720).load(f"{data_root}/gravity/static/GOCO05c.gfc")
```

```python
# Plot gravity anomalies
contour, cbar = plt.synthesis(goco05c, 1440, 720, GravityAnomaly())
contour.set_clim(-200, 200)
cbar.mappable = contour
cbar.set_ticks([-200, -100, 0, 100, 200])
cbar.ax.set_ylim(-200, 200)
plt.savefig("static-thin.png", dpi=600)
plt.show()
```

```python
# Plot geoid heights
plt.synthesis(goco05c, 720, 360, GeoidHeight())
plt.show()
```
