## GRACE normals processing

```python
# Imports
from gfeatpy.gravity import SphericalHarmonicsError, EquivalentWaterHeight
from gfeatpy import plot as plt
```

### Non-resonant case

```python
# Load monthly normal matrix
monthly_error = SphericalHarmonicsError(90, "gravity/monthly_normals/ITSG-Grace2018_n96_2007-03.snx")  # EXPENSIVE!! ~1 min
```

```python
# Plot EWH commission error
plt.synthesis(monthly_error, 180, 90, EquivalentWaterHeight(0))
plt.show()
```

```python
# Plot EWH + Gaussian spatial smoothing filter r=750 km
plt.synthesis(monthly_error, 180, 90, EquivalentWaterHeight(750e3))
plt.show()
```
